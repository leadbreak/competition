```python
# 공격 직전 상황 체크 1 - stride = 10

before_10_index = []

for i in range(len(df4)) :
  if (i > 10) and (df4['attack'][i] == 1) and (df4['attack'][i-1] == 0) :
    
    for j in range(i-10, i) :
      before_10_index.append(j)

before_10_df = df4.loc[before_10_index,:].reset_index(drop=True)
before_10_df
```

```python
# 공격 직전 상황 체크 2 - stride = 5

before_5_index = []

for i in range(len(df4)) :
  if (i > 5) and (df4['attack'][i] == 1) and (df4['attack'][i-1] == 0) :
    
    for j in range(i-5, i) :
      before_5_index.append(j)

before_5_df = df4.loc[before_5_index,:].reset_index(drop=True)
before_5_df
```

```python
# 공격 직전 상황 체크 3 - stride = 3

before_3_index = []

for i in range(len(df4)) :
  if (i > 3) and (df4['attack'][i] == 1) and (df4['attack'][i-1] == 0) :
    
    for j in range(i-3, i) :
      before_3_index.append(j)

before_3_df = df4.loc[before_3_index,:].reset_index(drop=True)
before_3_df
```
